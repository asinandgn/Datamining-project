Age Group OluÅturma

- age <= 19, teens
- 19 > age >= 39, adults
- 40 > age >= 59, middle age adults
- age >= 60, seniors


```python
cp.loc[ cp['Age'] <= 19  ,  "AgeGroup"  ] = "Teen"
cp.loc[ (cp['Age'] > 20) & (cp['Age'] <= 39)   ,  "AgeGroup"  ] = "Adults"
cp.loc[ (cp['Age'] > 40) & (cp['Age'] <= 59)  ,  "AgeGroup"  ] = "Middle Age Adults"
cp.loc[ (cp['Age'] >= 60)  ,  "AgeGroup"  ] = "Seniors"
```

```python
#KÃ¼tÃ¼phanenin eklenmesi ve verinin eÄitim ve test olarak ayrÄ±lmasÄ±
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(final2Cp, finalCp.Children, test_size=0.33, random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
#model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(x_train, y_train)
tahmin = knn.predict(x_test)
```

```python
#SeÃ§ilen modelde en uygun k deÄerinin bulunmasÄ± iÃ§in  bir dÃ¶ngÃ¼ ile oluÅrurulan listeye deÄerlerimizi atayalÄ±m
basari_list=[]
for each in range(2,12):
    knn1 = KNeighborsClassifier(n_neighbors=each)
    knn1.fit(x_train, y_train)
    basari_list.append(knn1.score(x_test, y_test))

plt.title("En uygun k")
plt.plot(range(2,12), basari_list)
plt.xlabel("k degeri")
plt.ylabel("basari")
plt.show()
```

```python
#GÃ¶rÃ¼ldÃ¼ÄÃ¼ Ã¼zere k=6 deÄeri ile en yÃ¼ksek tahminleme yapÄ±labilir.
#k deÄerinin artmasÄ±nÄ±n bir maliyet artÄ±ÅÄ±na neden olduÄu unutulmamalÄ±dÄ±r
```
