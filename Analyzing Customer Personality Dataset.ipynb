Age Group Oluşturma

- age <= 19, teens
- 19 > age >= 39, adults
- 40 > age >= 59, middle age adults
- age >= 60, seniors


```python
cp.loc[ cp['Age'] <= 19  ,  "AgeGroup"  ] = "Teen"
cp.loc[ (cp['Age'] > 20) & (cp['Age'] <= 39)   ,  "AgeGroup"  ] = "Adults"
cp.loc[ (cp['Age'] > 40) & (cp['Age'] <= 59)  ,  "AgeGroup"  ] = "Middle Age Adults"
cp.loc[ (cp['Age'] >= 60)  ,  "AgeGroup"  ] = "Seniors"
```

```python
#Kütüphanenin eklenmesi ve verinin eğitim ve test olarak ayrılması
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(final2Cp, finalCp.Children, test_size=0.33, random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
#model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(x_train, y_train)
tahmin = knn.predict(x_test)
```

```python
#Seçilen modelde en uygun k değerinin bulunması için  bir döngü ile oluşrurulan listeye değerlerimizi atayalım
basari_list=[]
for each in range(2,12):
    knn1 = KNeighborsClassifier(n_neighbors=each)
    knn1.fit(x_train, y_train)
    basari_list.append(knn1.score(x_test, y_test))

plt.title("En uygun k")
plt.plot(range(2,12), basari_list)
plt.xlabel("k degeri")
plt.ylabel("basari")
plt.show()
```

```python
#Görüldüğü üzere k=6 değeri ile en yüksek tahminleme yapılabilir.
#k değerinin artmasının bir maliyet artışına neden olduğu unutulmamalıdır
```

```python

#F1 score, bir sınıflandırma modelinin hassasiyet (precision) ve duyarlılık (recall) metriklerini birleştirerek modelin 
#genel performansını ölçen bir metriktir. F1 score, genellikle 0 ile 1 arasında bir değer alır. F1 score değeri ne olmalıdır 
#sorusu, genelde spesifik bir hedefe yönelik bir cevap vermek zordur. Ancak, F1 score'un değerlendirilmesi için genel bazı
#kılavuzlar şunlar olabilir:

#Yüksek Değerler: Yüksek bir F1 score (1'e yakın) genellikle modelin hem hassasiyeti hem de duyarlılığı yüksek bir şekilde 
#birleştirdiğini gösterir. Bu, modelin hem doğru pozitifleri ne kadar iyi tespit ettiğini hem de yanlış pozitifleri ne kadar 
#iyi önlediğini gösterir.

#Dengeli Değerler: Hassasiyet ve duyarlılık birbirine yakınsa, yani model hem pozitifleri doğru tespit edip hem de 
#negatifleri doğru reddediyorsa, F1 score yüksek olacaktır.

#Düşük Değerler: Düşük bir F1 score, modelin ya pozitifleri tespit etmede ya da negatifleri doğru bir şekilde sınıflandırmada
#zorlandığını gösterebilir. Modelin performansını iyileştirmek için dengesizliği ele almak veya modeli ayarlamak gerekebilir.
```

```python
#Bu örnekte, cohen_kappa_score fonksiyonu, gerçek sınıflar ve tahmin edilen sınıflar arasındaki kappa istatistiğini hesaplar.
#Fonksiyon, iki sınıf listesini alır ve kappa istatistiğini döndürür.
#Kappa değeri genellikle -1 ile 1 arasında bir değer alır. 1, mükemmel bir uyumu, 0, rastgele bir tahmini ve -1, 
#mükemmel bir uyumsuzluğu temsil eder. Kappa değeri 0'dan büyükse, modelin rastgele tahminlere göre daha iyi olduğunu 
#gösterir. Kappa istatistiği, özellikle dengesiz sınıf dağılımlarına sahip veri setlerinde doğruluk ölçütünü tamamlayıcı 
#olarak kullanılabilir. Ancak, veri setinin özelliklerine bağlı olarak farklı yorumlar yapılabilir, bu nedenle dikkatlice 
#değerlendirilmelidir.
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
y_olasılık = knn.predict_proba(x_test)[:, 1]
```

```python
# ROC eğrisini ve AUC'yi hesapla
fpr, tpr, thresholds = roc_curve(y_test, y_olasılık)
auc = roc_auc_score(y_test, y_olasılık)

# ROC eğrisini çiz
plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Rastgele')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Eğrisi')
plt.legend()
plt.show()
```

```python
#ROC eğrisi, farklı threshold değerlerinde TPR ve FPR'nin nasıl değiştiğini gösterir. Eğri altındaki alan 
#(Area Under the Curve - AUC), ROC eğrisinin altında kalan alanı temsil eder ve bu değer, modelin genel 
#sınıflandırma performansını ölçer. AUC değeri 1'e yaklaştıkça, modelin performansı iyileşir. AUC değeri 
#0.5'e yaklaşıyorsa, modelin performansı rastgele bir sınıflandırıcıya benzer.
```
